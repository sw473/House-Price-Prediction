# Building Linear Regression Model

```python
data.describe(include='all')
```

```python
data.head()
```

```python
data.drop(['PRT_ID'], axis=1, inplace = True)
```

```python
data = pd.get_dummies(data)
```

```python
x=data.drop('SALES_PRICE', axis=1)
```

```python
y= data['SALES_PRICE']
```

```python
from sklearn.model_selection import train_test_split as tts
train_x, test_x, train_y, test_y = tts(x, y, test_size = 0.3, random_state = 2)
```

```python
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_log_error as rmsle
```

```python
lr = LR()
lr.fit(train_x, train_y)
```

# Model evaluation using rmsle

```python
train_pred = lr.predict(train_x)
train_score = np.sqrt(rmsle(train_y,train_pred))
```

```python
test_pred = lr.predict(test_x)
test_score=np.sqrt(rmsle(test_y,test_pred))
```

```python
print('Training score:', train_score)
print('Test score:', test_score)
```
